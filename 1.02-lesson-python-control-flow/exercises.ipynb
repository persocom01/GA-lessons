## Exercise 2: Divisors
Given an integer, assemble a list of all its divisors.

```python
def find_divisors(n):
    divisors = []
    for i in range(2, int(n/2) + 1):
        if n%i == 0:
            divisors.append(i)
    print(divisors)
```

```python
find_divisors(88)
```
